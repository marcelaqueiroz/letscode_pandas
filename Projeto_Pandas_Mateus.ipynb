#### Etapa 1 : Juntando os Data Frames 

````python
#Alimentando o DataFrama a partir dos arquivos csv disponibilizados pelo Kuma
df_demografico = pd.read_csv('1_demografico.csv', sep =';')

df_renda_gastos = pd.read_csv('2_renda_gastos.csv', sep =';')

df_bens = pd.read_csv('3_bens.csv', sep =';')
````

````python
# Merge dataframes
data_frame_total = (df_demografico
               .merge(df_renda_gastos, how='inner', on=['ID', 'Region'])
               .merge(df_bens, how='inner', right_index=True, left_index=True))
data_frame_total
````

#### Etapa 3 : Tratamento de Outliers

Estabelecendo os limites inferior e superior para posterior tratamento de outliers

````python
lim_min = data_frame_total.quantile(0.25) - 1.5*(data_frame_total.quantile(0.75) - data_frame_total.quantile(0.25))
# print(lim_min)

lim_max = data_frame_total.quantile(0.75) + 1.5*(data_frame_total.quantile(0.75) - data_frame_total.quantile(0.25))
# print(lim_max)
````

````python
df_filtrado_total = data_frame_total[(data_frame_total <= lim_max) & (data_frame_total >= lim_min)].drop(columns = ['ID'])

df_filtrado_total = df_filtrado_total.select_dtypes(include = ['number'])
````

Geração do relatório com os outliers retirados

````python
df_filtrado_total.describe().T
````

#### Etapa 4 : Tratar os outliers e colocar a mediana no lugar, gerando o relatório de novo

````python
df_filtrado_total.fillna(data_frame_total.median(), inplace=True)

df_filtrado_total.describe().T
````

#### Etapa 5 : Relatório com variáveis qualitativas

````python
qualitativas = data_frame_total.select_dtypes(exclude = np.number)

qualitativas.head()
````

Preciso arrumar conforme o pessoal mostrou 

````python
for column in qualitativas:
    relatorio_qualitativas = pd.concat([
        pd.crosstab(index=data_frame_total[column], columns='Frequency'),
        pd.crosstab(index=data_frame_total[column], columns='Cumulative Frequency').cumsum(),
        (pd.crosstab(index=data_frame_total[column], columns='% Frequency', normalize='columns')*100).round(2),
        (pd.crosstab(index=data_frame_total[column], columns='Cumulative % Frequency', normalize='columns')*100).cumsum().round(2)],
        axis=1)
    relatorio_qualitativas.loc['Total'] = [relatorio_qualitativas['Frequency'].sum(),relatorio_qualitativas['Cumulative Frequency'][-1],\
                                relatorio_qualitativas['% Frequency'].sum(), relatorio_qualitativas['Cumulative % Frequency'][-1]]
    display(relatorio_qualitativas)
````

````python
relatorio_qualitativas = pd.DataFrame({
    'Freq Abs' : qualitativas.count(),
    'Freq Abs Acum' : qualitativas.count().cumsum(),
    'Freq Rel (%)' : (qualitativas.count()/qualitativas.size)*100,
    'Freq Rel Acum' : (qualitativas.count()/qualitativas.size).cumsum()
})

relatorio_qualitativas
````
