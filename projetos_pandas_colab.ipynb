```python
import numpy as np
import pandas as pd
```

```python
demografico = '/content/drive/MyDrive/Curso Santander/Modulo 2/Projeto_Pandas/1_demografico.csv'
renda = '/content/drive/MyDrive/Curso Santander/Modulo 2/Projeto_Pandas/2_renda_gastos.csv'
bens = '/content/drive/MyDrive/Curso Santander/Modulo 2/Projeto_Pandas/3_bens.csv'
```

```python
df_demografico =  pd.read_csv(demografico,sep=';')
df_renda = pd.read_csv(renda,sep=';')
df_bens = pd.read_csv(bens,sep=';')

```

```python
df_demografico.info()
```

```python
df_renda.info()
```

```python
base_1 = pd.merge(df_demografico,df_renda,how='outer')
#base_1.info()


```

```python
df = pd.merge(base_1,df_bens,how='outer',left_index=True,right_index=True)
```

```python
df.info()
```

```python
estatistica = df.describe()
df.describe()
```

###QuestÃ£o 3

```python
df_questao3 = df.copy()

filtro_colunas =  estatistica.columns

for i in filtro_colunas[1:]:
  iqr = df_questao3[i].quantile(0.75)-df_questao3[i].quantile(0.25)
  maximo = df_questao3[i].quantile(0.75)+1.5*iqr
  minimo = df_questao3[i].quantile(0.25)-1.5*iqr
  df_questao3.loc[df[i] > maximo,i] = np.nan
  df_questao3.loc[df[i]<minimo,i] =np.nan

df_questao3.describe()
```

###QuestÃ£o 4

```python
df_questao4 = df.copy()

filtro_colunas =  estatistica.columns

for i in filtro_colunas[1:]:
  iqr = df_questao4[i].quantile(0.75)-df_questao4[i].quantile(0.25)
  maximo = df_questao4[i].quantile(0.75)+1.5*iqr
  minimo = df_questao4[i].quantile(0.25)-1.5*iqr
  mediana = df_questao4[i].median()
  df_questao4.loc[df[i] > maximo,i] = mediana
  df_questao4.loc[df[i]<minimo,i] = mediana

df_questao4.describe()
```

###QuestÃ£o 5

```python
df_qualitativos = df.copy()
#df_qualitativos.select_dtypes(exclude=['number'])
a= np.array(df_qualitativos.dtypes != object)
b = np.array(df_qualitativos.columns)
filtro = b[a]
df_qualitativos.drop(filtro,axis=1,inplace=True)

```

```python
colunas_qualitativas =  df_qualitativos.columns
#df_qualitativos[colunas_qualitativas[1]].crosstab()
# relatorio = pd.crosstab(index=df_qualitativos[colunas_qualitativas[6]], columns='Frequencia')
# relatorio

for coluna in colunas_qualitativas:
   data_quali = pd.concat(
        [pd.crosstab(index=df_qualitativos[coluna], columns='Frequencia'),
         (pd.crosstab(index=df_qualitativos[coluna], columns='Freq. rel ', normalize='columns')).round(2),
         (pd.crosstab(index=df_qualitativos[coluna], columns='Freq. rel (%)', normalize='columns')*100).round(2),
         (pd.crosstab(index=df_qualitativos[coluna], columns='Freq. rel. Acu', normalize='columns')).cumsum().round(2)],
         axis=1)
   data_quali.loc['TOTAL'] = data_quali.sum()
   display(data_quali)
```
